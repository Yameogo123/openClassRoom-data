```python
def _(text):
    punctuation_signs = ['!', '(', ')', ',', '-', '.',':', ';', '?','_']
    tokens= WordPunctTokenizer().tokenize(text)
    tokens=[x for x in tokens if x not in punctuation_signs]
    return len(tokens)

L=[_(sen) for sen in sentences]
np.mean(L)
```

```python
doc = nlp(text)
```

```python
sentences = [sent.text.replace('\r\n', ' ').strip() for sent in doc.sents]
sentences
```

```python
def extract_pos(pos_type, sentences):
    lemmas = []
    for sentence in sentences:
        sdoc = nlp(sentence)
        lemmas += [token.lemma_ for token in sdoc if token.pos_ == pos_type]
    return lemmas
```

```python
lemmas = extract_pos('VERB', sentences)
Counter(lemmas).most_common(10)
```

```python
lemmas = extract_pos('PRON', sentences)
Counter(lemmas).most_common(30)
```

```python
lemmas = extract_pos('PROPN', sentences)
Counter(lemmas).most_common(10)
```

```python
lemmas = extract_pos('ADJ', sentences)
Counter(lemmas).most_common(10)
```
