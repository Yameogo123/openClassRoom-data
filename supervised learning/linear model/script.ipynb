```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df= pd.read_csv("dataset_feuille.csv")
df
```

```python
#les elements null
na=df.isna().sum().sum()
na
```

```python
#doublons
dbl= df.duplicated()
doublon=df[dbl].sum().sum()
doublon
```

```python
#unicité
uni=df.nunique()
uni[uni==1]
```

```python
if "id" in df.columns:
    df.drop(columns=["id"], inplace=True)
```

```python
y= df.species
x= df.iloc[:, 1:]
scaler= StandardScaler()
x= scaler.fit_transform(x)
```

```python
lb= LabelEncoder()
fitlb= lb.fit(y)
y= fitlb.transform(y)
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(x, y, test_size=0.25)
```

```python
svm= LinearSVC()
lr= LogisticRegression(C=0.9102981779915218)
```

```python
#svm
svm.fit(xtrain, ytrain)
proba= svm.decision_function(xtest)
score=svm.score(xtest, ytest)
score
```

```python
#regression logistic
lr.fit(xtrain, ytrain)
score2= lr.score(xtest, ytest)
score2
```

```python
params= {
    "C": np.logspace(-2, 2, 10), 
    "penalty":["11", "12"],
    "solver":["lbfgs", "newton-cg", "liblinear"]
}
```

```python
grid= GridSearchCV(lr, params, cv=10, n_jobs=-1, return_train_score=True, verbose=1)
grid.fit(xtrain, ytrain)
grid.best_params_
```

```python
test= pd.read_csv("test.csv")
X= test.iloc[:, 1:]
X=scaler.fit_transform(X)
```

```python
ypred=lr.predict(X)
species=lb.inverse_transform(ypred)
```

```python
test["species"]=species
```
